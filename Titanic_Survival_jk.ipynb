```python
import pandas as pd
import numpy as np
```

Reading data using **pandas**

```python
df = pd.DataFrame(pd.read_csv('/content/train.csv'))
df.head()
```

```python
df.shape
```

**Handeling NULL values**

```python
df.isnull().sum()
```

seprating out th columns have more than 35% of the values missing in the dataset

```python
drop_col = df.isnull().sum()[df.isnull().sum()>(35/100 * df.shape[0])]
drop_col
```

```python
drop_col.index
```

```python
df.drop(drop_col.index, axis=1, inplace=True)
df.isnull().sum()
```

```python
df.fillna(df.mean(), inplace = True)
df.isnull().sum()
```

**Embrked**

```python
df['Embarked'].describe()
```

now fill the **NULL** values in Embarked attribute

```python
df['Embarked'].fillna('S', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.corr()
```

**sibsp and parch** are combining in a new column

```python
df['FamilySize'] = df['SibSp']+df['Parch']
df.drop(['SibSp', 'Parch'], axis=1, inplace=True)
df.corr()
```

FamilySize in the ship does not have much **correlance** with survival rate

Let's check if the person was alone or not can affect the survival rate

```python
df['Alone'] = [0 if df['FamilySize'][i]>0 else 1 for i in df.index]
df.head()
```

```python
df.groupby(['Alone'])['Survived'].mean()
```

if the person is alone has less chance of surviving

```python
df[['Alone', 'Fare']].corr()
```

if the person was not alone the chance the ticket price is higher

```python
df['Sex'] = [0 if df['Sex'][i]=='male' else 1 for i in df.index]
df.groupby(['Sex'])['Survived'].mean()
```

female paassengers have more chance of survive than males one

```python
df.groupby(['Embarked'])['Survived'].mean()
```
